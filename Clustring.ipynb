```python
# importing Library 
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
from scipy.cluster.hierarchy import linkage, dendrogram


# Path to get files
current_dir = os.getcwd()
relative_path = "Study_2_data/result_X201SC22101900-Z01-F019/02.FeatureAnalysis/tables/sample/featureTable_Relative"
data_path = os.path.join(current_dir,relative_path)
print(data_path)
```

```python

# Data Load 
# //==============================================

Phylum_file = "featureTable.sample.p.relative.xls"
Genus_file =  "featureTable.sample.g.relative.xls"
Species_file = "featureTable.sample.s.relative.xls"

# Load files using pandas
phylum_data = pd.read_table(os.path.join(data_path, Phylum_file),sep="\t")
genus_data = pd.read_table(os.path.join(data_path, Genus_file),sep="\t")
species_data = pd.read_table(os.path.join(data_path, Species_file),sep="\t")
```

```python
# drop the Tax_detail column from phylum data
phylum_data = phylum_data.drop('Tax_detail', axis=1)

phylum_data

# Extract unique taxonomy labels
unique_taxonomy_data = phylum_data['Taxonomy'].unique()
unique_taxonomy_data = [str(x) for x in unique_taxonomy_data]

# Traspose the Phylum data
#phylum_transposed = pd.DataFrame(phylum_data.T)

# set index 
phylum_data = phylum_data.set_index('Taxonomy')

# Perform hierarchical clustering
sns.clustermap(phylum_data,metric="euclidean",method='ward',cmap="Blues")
plt.show()
```
