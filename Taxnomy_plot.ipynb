```python
# importing Library 
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 

# Path to get files
current_dir = os.getcwd()
relative_path = "Study_2_data/result_X201SC22101900-Z01-F019/02.FeatureAnalysis/tables/sample/featureTable_Relative"
data_path = os.path.join(current_dir,relative_path)
print(data_path)
```

```python
# Data Load 
# //==============================================

Phylum_file = "featureTable.sample.p.relative.xls"
Genus_file =  "featureTable.sample.g.relative.xls"
Species_file = "featureTable.sample.s.relative.xls"

# Load files using pandas
phylum_data = pd.read_table(os.path.join(data_path, Phylum_file),sep="\t")
genus_data = pd.read_table(os.path.join(data_path, Genus_file),sep="\t")
species_data = pd.read_table(os.path.join(data_path, Species_file),sep="\t")

phylum_data
```

```python
# Assuming taxonomy_colors is defined outside the function, similar to your previous code
# taxonomy_colors = dict(zip(taxonomies, colormap))
import seaborn as sns

taxonomies = ['TaxonomyA', 'TaxonomyB', 'TaxonomyC']
color_palette = sns.color_palette("viridis", len(taxonomies))

taxonomy_colors = dict(zip(taxonomies, color_palette))


def create_pie_plot(sample_id, data, column_name, threshold=0.05, level_colors=None):
    # Filter data for the selected sample and abundance over the threshold
    filtered_data = data[(data['Sample'] == sample_id) & (data[column_name] > threshold)]

    # Resetting the index to avoid potential KeyError
    filtered_data.reset_index(drop=True, inplace=True)
    
    # Create a pie plot using the "Percentage" column
    plt.figure(figsize=(6, 6))

    # Create the pie plot with explode
    if level_colors is None:
        level_colors = taxonomy_colors  # Use the default taxonomy colors if not provided

    colors = [level_colors[taxon] for taxon in filtered_data['Taxonomy']]
    patches, autotexts, text = plt.pie(x=filtered_data[column_name], pctdistance=0.5, autopct=lambda p: '{:.1f}%'.format(p) if p >= 3 else '',
                                       colors=colors)

    # add Legends
    legend_labels = [f"{taxonomy} - {percentage:.1f}%" for taxonomy, percentage in zip(filtered_data['Taxonomy'], filtered_data[column_name])]
    plt.legend(legend_labels, title='Taxonomy - % Abundance', loc='center left', bbox_to_anchor=(1, 0.5),fontsize='x-small')

    # Update layout
    plt.title(f"Pie Plot for Sample: {sample_id}")
    plt.tight_layout()

    # Show the plot
    plt.show()

```

```python
# Taxonomy plot at Phylum level 
# //=======================================
phylum_Taxa = phylum_data['Taxonomy']
color_palette = sns.color_palette("husl", len(phylum_Taxa))
taxonomy_colors = dict(zip(phylum_Taxa, color_palette))



phylum_data_melted = pd.melt(phylum_data,id_vars=['Taxonomy','Tax_detail'], var_name='Sample', value_name='Abundance_Value')

# Calculate percentage of abundance for each individual sample
phylum_data_melted['Abundance_Percentage'] = phylum_data_melted.groupby('Sample')['Abundance_Value'].transform(lambda x: (x / x.sum()) * 100)

# Plot pie charts for each sample
unique_samples_p = phylum_data_melted['Sample'].unique()

for sample in unique_samples_p:
    create_pie_plot(sample,phylum_data_melted,'Abundance_Percentage')
```

```python
# Taxonomy plot at Species level 
# //=======================================

species_Taxa = species_data['Taxonomy']
color_palette = sns.color_palette("Paired", len(species_Taxa))
taxonomy_colors = dict(zip(species_Taxa, color_palette))

Species_data_melted = pd.melt(species_data,id_vars=['Taxonomy','Tax_detail'], var_name='Sample', value_name='Abundance_Value')

# Calculate percentage of abundance for each individual sample
Species_data_melted['Abundance_Percentage'] = Species_data_melted.groupby('Sample')['Abundance_Value'].transform(lambda x: (x / x.sum()) * 100)

# Plot pie charts for each sample
unique_samples_s = Species_data_melted['Sample'].unique()

for sample in unique_samples_s:
    create_pie_plot(sample,Species_data_melted,'Abundance_Percentage')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assuming phylum_data is your data frame with the necessary structure

Phyla = phylum_data.iloc[:, 0].values

# Get total abundances
abundances = phylum_data.iloc[:, 1:].sum(axis=1).values

# Create dataframe
Phyla_total_abundance = pd.DataFrame({'Phylum': Phyla, 'Abundance': abundances})

# Filter abundances
Phyla_total_abundance = Phyla_total_abundance[Phyla_total_abundance['Abundance'] >= 0.01]

# adding Explode
explode = []
for i, row in Phyla_total_abundance.iterrows():
    if row['Abundance'] < 1:
        explode.append(0.1)
    else:
        explode.append(0)


# Plot pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(Phyla_total_abundance['Abundance'], autopct= lambda p: '{:.1f}%'.format(p) if p >= 1 else '', explode=explode)


# Get total abundance 
total = Phyla_total_abundance['Abundance'].sum()

# Calculate percentages
percentages = Phyla_total_abundance['Abundance'] / total * 100

# Create labels
# Create labels
legend_labels = [f"{taxonomy} ({p:.2f}%)" for taxonomy, p in zip(Phyla_total_abundance['Phylum'], percentages)]
plt.legend(legend_labels, title='Phylum (Abundance)', loc='center left', bbox_to_anchor=(1, 0.5),frameon= False)
plt.title("Phylum Abundance")
plt.tight_layout()
plt.show()
```
