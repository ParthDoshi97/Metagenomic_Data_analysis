```python
# importing Library 
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 

# Path to get files
current_dir = os.getcwd()
relative_path = "Study_2_data/result_X201SC22101900-Z01-F019/04.AlphaDiversity/sample/alpha_index_table"
alfa_path = os.path.join(current_dir,relative_path)
print(alfa_path)
```

```python

# load Data 
Alfa_diversity = pd.read_csv(os.path.join(alfa_path,"alpha_diversity_index.sample.txt"), delimiter="\t")

Alfa_diversity = pd.DataFrame(Alfa_diversity)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your data is stored in a DataFrame named 'Alfa_diversity'
# Columns: 'Sample_Name', 'Index', 'Value'

# Indices to compare
indices_to_compare = ['chao1', 'dominance', 'goods_coverage', 'observed_features', 'pielou_e', 'shannon', 'simpson']

# Melt the DataFrame to a long format for seaborn
melted_data = pd.melt(Alfa_diversity, id_vars=['Sample_Name'], value_vars=indices_to_compare, var_name='Index', value_name='Value')

# Set the style for the plot (optional)
sns.set(style="white")

# Get unique indices
indices_to_compare = melted_data['Index'].unique()

# Create a color palette for the dot plots
color_palette = sns.color_palette("Set2", n_colors=len(indices_to_compare))

# Create a 2x2 grid of dot plots for alpha diversity indices
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(15, 10))

# Flatten the 2x2 grid for easy indexing
axes = axes.flatten()

for i, index in enumerate(indices_to_compare):
    color = color_palette[i]
    sns.stripplot(x='Sample_Name', y='Value', data=melted_data[melted_data['Index'] == index], color=color, jitter=True, alpha=0.7, ax=axes[i])
    
    # Customize each subplot
    axes[i].set_ylabel(f'{index} Value')
    axes[i].set_xlabel('Sample_Name')
    axes[i].grid(True, axis='y', linestyle='--', alpha=0.7)

    # Rotate x-axis labels
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=90, ha='right')

    # Remove the empty subplot at the last position
fig.delaxes(axes[-1])
# Adjust layout and show the plot
plt.tight_layout()
plt.show()

```
